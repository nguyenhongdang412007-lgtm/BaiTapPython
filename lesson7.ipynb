```python
#w7a1
def BS(nums,target,l,r):
    if l > r: return -1
    m=(l+r)//2
    if (nums[m]==target): return m
    elif (nums[m]>target): return BS(nums,target,l,m-1)
    else: return BS(nums,target,m+1,r)
nums=list(map(int,input().split()))
target=int(input())
print(BS(nums,target,0,len(nums)-1))
```

```python
#w7a2
def countt(arr,x):
    bool_x=False
    d={}
    for i in arr:
        if i==x: bool_x=True
        if i in d:
            d[i]+=1
        else:
            d[i]=1
    return d[x] if bool_x else 0
arr=list(map(int,input().split()))
x=int(input())
print(countt(arr,x))
```

```python
#w7a3
nums=list(map(int,input().split()))
def bubbleSort(nums):
    n=len(nums)
    for i in range(n):
        for j in range(n-i-1):
            if (nums[j]>nums[j+1]):
                nums[j],nums[j+1]=nums[j+1],nums[j]
bubbleSort(nums)
print(nums)
```

```python
nums=list(map(int,input().split()))
def selectionSort(nums):
    n=len(nums)
    for i in range(n):
        key=i
        for j in range(i+1,n):
            if nums[j]<nums[key]:
                key=j
        if i!=key: nums[i],nums[key]=nums[key],nums[i]
selectionSort(nums)
print(nums)
```

```python
#w7a5
def countsum(nums,target):
    d={}
    res=0
    for i in range(len(nums)):
        k=target-nums[i]
        if k in d:
            res+=d[k]
        if nums[i] in d:
            d[nums[i]]+=1
        else:
            d[nums[i]]=1
    return res
nums=list(map(int,input().split()))
target=int(input())
print(countsum(nums,target))
```

```python
#w7a8
def lower_bound(nums,target,l,r):
    if l>r: return l
    m=(l+r)//2
    if(nums[m]>=target): return lower_bound(nums,target,l,m-1)
    else: return lower_bound(nums,target,m+1,r)
def upper_bound(nums,target,l,r):
    if l>r: return l
    m=(l+r)//2
    if (target<nums[m]): return upper_bound(nums,target,l,m-1)
    else: return upper_bound(nums,target,m+1,r)
def findMinIterval(intervals,queries):
    intervals.sort(key=lambda x:x[1]-x[0])
    n=len(queries)
    values_queries=[]
    for i in range(n):
        values_queries.append([queries[i],i])
    values_queries.sort(key=lambda x:x[0])
    queries.sort()
    res=[-1]*n
    for i in intervals:
        l=lower_bound(queries,i[0],0,n-1)
        r=upper_bound(queries,i[1],0,n-1)
        for j in range(l,r):
            validQ=values_queries[j]
            idx=validQ[1]
            if (res[idx]==-1):
                res[idx]=i[1]-i[0]+1
    return res
m=int(input())
interval=[list(map(int,input().split())) for _ in range(m)]
queries=list(map(int,input().split()))
print(findMinIterval(interval,queries))
```

```python
#w7a9
def merge(nums,res,idx,l,m,r):
    i,j=l,m+1
    arr=[]
    rCount=0
    while(i<m+1 and j<r+1):
        if(nums[idx[i]]<=nums[idx[j]]):
            res[idx[i]]+=rCount
            arr.append(idx[i])
            i+=1
        else:
            rCount+=1
            arr.append(idx[j])
            j+=1
    while(i<m+1):
        res[idx[i]]+=rCount
        arr.append(idx[i])
        i+=1
    while(j<r+1):
        arr.append(idx[j])
        j+=1
    for k in range(len(arr)):
        idx[l+k]=arr[k]
def mergeSort(nums,res,idx,l,r):
    if l>=r: return
    m=(l+r)//2
    mergeSort(nums,res,idx,l,m)
    mergeSort(nums,res,idx,m+1,r)
    merge(nums,res,idx,l,m,r)
def countL(nums):
    n=len(nums)
    res=[0]*n
    idx=[0]*n
    for i in range(n): idx[i]=i
    mergeSort(nums,res,idx,0,n-1)
    return res
nums=list(map(int,input().split()))
print(countL(nums))
```

```python
#w7a10
def upper_bound(nums,target,l,r):
    if l>r: return l
    m=(l+r)//2
    if (target<nums[m]): return upper_bound(nums,target,l,m-1)
    else: return upper_bound(nums,target,m+1,r)
def moduloM(nums,m):
    pre=0
    res=0
    presum=set()
    for i in range(len(nums)):
        k=list(presum)
        pre+=nums[i]
        mod=pre%m
        res=max(res,mod)
        idx=upper_bound(k,mod,0,len(k)-1)
        if (idx!=len(k)): res=max(res,mod-k[idx]+m)
        presum.add(mod)
    return res
nums=list(map(int,input().split()))
m=int(input())
print(moduloM(nums,m))
```

```python
#w7a13
def lhs(nums):
    res=0
    d={}
    for i in range(len(nums)):
        if nums[i] in d:
            d[nums[i]]+=1
        else:
            d[nums[i]]=1
    for i in d:
        if i+1 in d:
            res=max(res,d[i]+d[i+1])
    return res
nums=list(map(int,input().split()))
print(lhs(nums))
```

```python
class Point3:
    def __init__(self,x,y,z):
        self.x=x
        self.y=y
        self.z=z
    def translate(self,newx,newy,newz):
        self.x+=newx
        self.y+=newy
        self.z+=newz
    def __str__(self):
        return f"({self.x},{self.y},{self.z})"
p = Point3(1,2,3)
p.translate(10, 0, -1)

assert (p.x, p.y, p.z) == (11,2,2)

print(str(p))
```
