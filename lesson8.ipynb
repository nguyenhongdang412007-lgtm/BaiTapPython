```python
from math import pi
class Cylinder:
  def __init__(self,r,h):
    self.radius = r
    self.height = h
  def area(self):
    return (2 * (self.radius **2) * pi) + (2 * self.radius * pi * self.height )
  def  v(self):
    return  (self.radius ** 2) * pi * self.height

r, h = map(int,input().split())
n = Cylinder(r, h)
print(n.area())
print(n.v())
```

```python
class Date:
  def __init__(self,d,m,y):
    self.day = d
    self.month = m
    self.year = y


  def _typeYear(self):
    if (self.year % 400 == 0) or (self.year % 100 != 0 and self.year % 4 == 0):
      self.calendar = [31,29,31,30,31,30,31,31,30,31,30,31]
    else:
      self.calendar = [31,28,31,30,31,30,31,31,30,31,30,31]

  def __str__(self):
    return "What's year?"

  def is_true(self):
    if self.month < 1 or self.month > 12:
      return False
    if self.day < 1 or self.day > self.calendar[self.month - 1]:
      return False
    return True

  def next_day(self):
    self.day += 1
    if self.day > self.calendar[self.month - 1]:
      self.day = 1
      self.month += 1
      if self.month > 12:
        self.month = 1
        self.year += 1
d,m,y = map(int,input().split("/"))
n = Date(d,m,y)
n._typeYear()
if n.is_true():
  n.next_day()
  print(f"Next day: {n.day}/{n.month}/{n.year}")
else:
  print("INVALID")
```

```python
import math
class Point:
  def __init__(self,x,y):
    self.x = x
    self.y = y
  def is_O(self):
    return True if (self.x == 0 and self.y == 0) else False
  def is_Ox(self):
    return True if (self.y == 0) else False
  def is_Oy(self):
    return True if (self.x == 0) else False
  def distance(self,other):
    return math.sqrt((self.x ** 2 - other.x ** 2) + (self.y ** 2 - other.y ** 2))

x,y = map(int,input("Nhập M(x,y): ").split())
M = Point(x,y)
O = Point(0,0)
if M.is_O():
  print("Là gốc tọa độ")
if M.is_Ox():
  print("Nằm ở trục hoành")
if M.is_Oy():
  print("Nằm ở trục tung")
print("Khoảng cách từ gốc tọa độ: " + f"{M.distance(O)}")
```

```python
class ShoppingCart:
  def __init__(self,items = None):
    self.items = {} if items == None else items
  def add(self,name, price):
    self.items[name] = price
  def remove(self, name):
    self.items.pop(name)
  def is_None(self):
    return True if len(self.items) == 0 else False
  def total(self):
    return sum(x for x in self.items.values())
  def see_cart(self):
    for x, y in self.items.items():
      print(f"{x:<20} Giá: {y} VND")
  def remove_all(self):
    self.items.clear()

cart = ShoppingCart()
running = True
while running:
  print("""
  ==============ShoppingCart=============
  1.Thêm vào giỏ
  2.Xóa khỏi giỏ
  3.Kiểm tra giỏ
  4.Tổng tiền hiện tại
  5.Xem giỏ
  6.Xóa toàn bộ giỏ
  7.Thoát
  """
  )
  n = int(input("Nhập yêu cầu của bạn: "))
  if n == 1:
    name = input("Nhập tên món đồ: ")
    price = int(input("Nhập giá của món đồ: "))
    cart.add(name, price)
  elif n == 2:
    name = input("Nhập tên món đồ xóa: ")
    cart.remove(name)
  elif n == 3:
    print("Giỏ hàng của bạn đang trống!" if cart.is_None() else "Giỏ hàng vẫn còn đồ!")
  elif n == 4:
    print(f"Tổng thanh toán của bạn: {cart.total()} VND")
  elif n == 5:
    print("================Giỏ hàng==============")
    cart.see_cart()
  elif n == 6:
    cart.remove_all()
    print("Xóa toàn bộ!")
  elif n == 7:
    print("Bạn chắc chắn muốn thoát ?(toàn bộ giỏ hàng sẽ bị xóa):")
    n = input("Tôi chắc chắn(y/n): ")
    if n == "y":
      running = False
    else:
      pass
  print("Hẹn gặp lại")

```

```python
class Rectangle:
  def __init__(self,w,h):
    self.w = w
    self.h = h
  def area(self):
    return self.w * self.h
  def perimeter(self):
    return 2 * ( self.w + self.h )
  def scale(self,k):
    self.w = self.w * k
    self.h = self.h * k

w, h , k = map(int,input().split())
I = Rectangle(w,h)
I.scale(k)
print(f"Diện tích {I.area()}, Chu vi {I.perimeter()}")
```

```python
from math import gcd
class Fraction:
  def __init__(self,num,den):
    self.num = num
    self.den = den
  def add(self,other):
    self.num = self.num * other.den + other.num * self.den
    self.den = self.den * other.den
  def sub(self,other):
    self.num = self.num * other.den - other.num * self.den
    self.den = self.den * other.den
  def mul(self, other):
    self.num *= other.num
    self.den *= other.den
  def div(self,other):
    self.num *= other.den
    self.den *= other.num
  def real_fraction(self):
    n = gcd(self.num,self.den)
    self.num //= n
    self.den //= n

  def __str__(self):
    return f"{self.num}/{self.den}"

n = input("Nhập yêu cầu (a b op c d): ")
a, b, op, c, d = n.split()
a, b, c, d = map(int, (a, b, c, d))

A = Fraction(a, b)
B = Fraction(c, d)
if op == "+":
  A.add(B)
elif op == "-":
  A.sub(B)
elif op == "*":
  A.mul(B)
elif op == "/":
  A.div(B)

A.real_fraction()
print(A)
```

```python
class BankAccount:
  def __init__(self,o,b):
    self.owner = o
    self.balance = b
  def deposit(self,money):
    self.balance += money
  def withdraw(self,money):
    if self.balance < money:
      print("Số dư không đủ")
    else:
      self.balance -= money
  def __str__(self):
    return f"Tài khoản {self.owner} - số dư : {self.balance}"

o, b = input("Nhập: ").split()
n = int(input("Nhập số lần: "))

b = float(b)
I = BankAccount(o,b)
while n > 0:
  f, m = input("Bạn muốn rút hay nạp tiền bao nhiêu?: ").split()
  m = float(m)
  if f.lower() == "deposit":
    I.deposit(m)
  elif f.lower() == "withdraw":
    I.withdraw(m)
  else:
    print("Vui lòng nhập đúng!")
    n += 1
  n -= 1
print(I)
```

```python
class Student:
  def __init__(self,name,scoreLst = None,scoreName = None):
    self.name = name
    self.scoreLst = [] if scoreLst is None else scoreLst
    self.scoreName = [] if scoreName is None else scoreName
  def add_score(self,subject, score):
    if subject not in self.scoreName:
      self.scoreLst.append(score)
      self.scoreName.append(subject)
  def average(self):
    return (sum(self.scoreLst) / len(self.scoreName))
  def Rank(self,avg):
    Rank = {
        5 : "Poor",
        6.5 : "Average",
        8 : "Good",
        10 : "Excellent",
    }
    for i,j in Rank.items():
      if avg < i:
        return j

name = input("Nhập tên:")
n = int(input("Số lần:"))
A = Student(name)
for i in range(n):
  a, b = input("Nhập môn / điểm : ").split()
  b = float(b)
  A.add_score(a,b)
print(f"{A.name} : {A.average():.2f} rank {A.Rank(A.average())}")


```

```python
class Book:
  def __init__(self, title, author, year):
    self.title = title
    self.author = author
    self.year = year
  def __str__(self):
    return f"{self.title} của tác giả {self.author} xuất bản năm {self.year}"

class Library:
  def __init__(self,Storage = None):
    self.storage = [] if Storage is None else Storage
  def add(self,book):
    self.storage.append(book)
  def count_by_author(self,author):
    count = 0
    for i in self.storage:
      if i.author == author:
        count += 1
    return count
  def find_by_year(self, year):
    count = 0
    for i in self.storage:
      if i.year == year:
        count += 1
    return count

I = Library()
repeat = int(input("Nhập số lần: "))
for i in range(repeat):
  func,book = input().split()
  if func == "ADD":
    title,author,year = book.split(";")
    book = Book(title,author,year)
    I.add(book)
  elif func == "COUNT":
    print(f"{I.count_by_author(book)}")
  elif func == "COUNTYEAR":
    print(f"{I.find_by_year(book)}")
  else:
    print("Nhập sai")

```

```python
#Muốn tạo ra không gian vector R^3 - Tức là hệ trục tọa độ Oxyz
#Với toàn bộ phép toán, class con tương ứng ( vector, mặt phẳng, điểm )
import math

class Oxyz:
  def isVector(self,point1,point2):
    pass
  def isflat(self,point1,point2,point3):
    pass


class Vector(Oxyz):
  def __init__(self,x,y,z):
    self.x = x
    self.y = y
    self.z = z
  def __str__(self):
    return f"({self.x};{self.y};{self.z})"
  def __add__(self,other):
    x = self.x + other.x
    y = self.y + other.y
    z = self.z + other.z
    return Vector(x,y,z)
  def __sub__(self, other):
    x = self.x - other.x
    y = self.y - other.y
    z = self.z - other.z
    return Vector(x,y,z)

  def __mul__(self,other):
    if type(other) == int or type(other) == float:
      return Vector(self.x * other, self.y * other, self.z * other)
    return self.x * other.x + self.y * other.y + self.z * other.z

  def __eq__(self,other):
    return True if (self.x == other.x and self.y == other.y and self.z == other.z) else False
  def is_parallel(self,other):
    return True if (math.lcm(self.x,other.x) == math.lcm(self.y,other.y) == math.lcm(self.z,other.z)) else False
  def angle(self,other):
    n = self * other
    cos_theta = n / (self.length() * other.length())
    PHI = max(-1,min(cos_theta,1))
    angle = math.acos(PHI)
    return math.degrees(angle)
  def length(self):
      return math.sqrt(self.x ** 2 + self.y ** 2 + self.z ** 2)
  def mulvector(self,other):
    x = self.y * other.z - self.z * other.y
    y = self.z * other.x - self.x * other.z
    z = self.x * other.y - self.y * other.x
    return Vector(x,y,z)

class Point(Oxyz):

  def __init__(self,x,y,z):
    self.x = x
    self.y = y
    self.z = z
  def __str__(self):
    return f"({self.x},{self.y},{self.z})"
  def distance(self, other):
    return math.sqrt((self.x - other.x) ** 2 + (self.y - other.y) ** 2 + (self.z - other.z) ** 2)
  def poisition_trans(self,vector):
    self.x += vector.x
    self.y += vector.y
    self.z += vector.z



class Plane(Oxyz):
  def __init__(self,point1,point2,point3):
    self.vector1 = Vector(point1.x - point2.x, point1.y - point2.y, point1.z - point2.z)
    self.vector2 = Vector(point1.x - point3.x, point1.y - point3.y, point1.z - point3.z)
    self.point1 = point1
    self.point2 = point2
    self.point3 = point3
  def plane_func(self):
    #Phương trình mặt phẳng có dạng Ax + By + Cz + D = 0
    #Với một point thuộc pt thì thỏa mãn pt trên
    #Suy ra D = - (A.xM + B.yM + C.zM)
    self.P_vector = self.vector1.mulvector(self.vector2)
    self.D= -(self.P_vector * self.point3)
  def distant_to_point(self,point):
    up = abs(self.P_vector * point + self.D)
    down = math.sqrt(self.P_vector.x ** 2 + self.P_vector.y ** 2 + self.P_vector.z ** 2)
    return up / down

a,b,c = map(float,input().split())
A = Point(a,b,c)
a2,b2,c2 = map(float,input().split())
B = Point(a2,b2,c2)
a3,b3,c3 = map(float,input().split())
C = Point(a3,b3,c3)
a4,b4,c4 = map(float,input().split())
D = Point(a4,b4,c4)

M = Plane(B,C,D)
M.plane_func()
vector_BC = Vector(B.x - C.x, B.y - C.y, B.z - C.z)
vector_AB = Vector(A.x - B.x, A.y - B.y, A.z - B.z)
dis_point_to_Line = (vector_BC.mulvector(vector_AB)).length() / vector_BC.length()
d = M.distant_to_point(A)
sin_PHI = d / dis_point_to_Line
PHI = max(-1, min(sin_PHI, 1))
angle = math.asin(PHI)
print(math.degrees(angle))
```

```python
class ComplexNumber:
  def __init__(self, real, imaginary):
    self.realPart = real
    self.imaginaryPart = imaginary
  def __mul__(self,other):
    a, b = self.realPart, self.imaginaryPart
    c, d = other.realPart, other.imaginaryPart
    new_realPart = a * c - b * d
    new_imaginaryPart = a * d + b * c
    return ComplexNumber(new_realPart,new_imaginaryPart)
  def __str__(self):
    return f"{self.realPart} + {self.imaginaryPart}i"

a, b = map(float, input("Nhập số thứ nhất: ").split())
c, d = map(float, input("Nhập số thứ hai: ").split())
A = ComplexNumber(a,b)
B = ComplexNumber(c,d)
print(A * B)
```

```python
from math import sqrt
class Triangle:
  def __init__(self,Edge1,Edge2,Edge3):
    self.edge = [Edge1,Edge2,Edge3]
  def getArea(self):
    p = sum(self.edge) / 2
    return sqrt(p * (p - self.edge[0]) * (p - self.edge[1]) * (p - self.edge[2]))
  def isTriangle(self):
    if 2 * max(self.edge) < sum(self.edge):
      return True
    return False

a, b, c = map(float,input("Nhập: ").split())
T = Triangle(a,b,c)
if T.isTriangle():
  print(f"{T.getArea():.2f}")
else:
  print(-1)
```
