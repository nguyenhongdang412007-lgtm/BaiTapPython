```python
def reverse(path: str) -> int:
    try:
        with open(path, 'r', encoding='utf-8') as f:
            n = f.read().strip()

        if not n.isdigit():
            raise ValueError("Nội dung tệp không phải số nguyên dương")

        # Đảo chuỗi rồi chuyển về int (tự bỏ số 0 ở đầu)
        return int(n[::-1])

    except FileNotFoundError:
        print("Không tìm thấy tệp")
        return -1
    except Exception as e:
        print("Lỗi:", e)
        return -1
```

```python
def canWinNim(path: str) -> bool:
    try:
        with open(path, 'r', encoding='utf-8') as f:
            n = int(f.read().strip())

        if n <= 0:
            raise ValueError("Số đá phải là số nguyên dương")

        return n % 4 != 0

    except FileNotFoundError:
        print("Không tìm thấy tệp")
        return False
    except Exception as e:
        print("Lỗi:", e)
        return False

```

```python
import os
path = input()
def find(path: str):
    if not os.path.exists(path):
        return ("Mission failed")
    max_val = -float('inf')
    min_val = float('inf')
    with open(path, "r") as f:
        for line in f:
            lst = line.split()
            for i in lst:
                if int(i) > max_val:
                    max_val = int(i)
                if int(i) < min_val:
                    min_val = int(i)
    return (max_val,min_val)
ma,mi = find(path)
print(ma)
print(mi)
```

```python
path = input()
def maximumProduct(path:str) -> int:
  with open(path, "r") as f:
    n = int(f.readline())
    line = f.readline()
    lst = list(map(int,line.split()))
    for i in range(1,len(lst)):
      key = lst[i]
      j = i - 1
      while lst[j] > key and j >=0:
        lst[j+1] = lst[j]
        j -= 1
      lst[j + 1] = key

  return lst[-1] * lst[-2] * lst[-3]

print(maximumProduct(path))


```

```python
def moveZeroes(fileName : str) -> list:
    with open(fileName, "r") as f:
        n = f.readline()
        line = f.readline()
        lst = list(map(int,line.split()))
    pivot = 0
    for i in range(len(lst)):
        if lst[i] != 0:
            lst[pivot] = lst[i]
            pivot += 1
    for i in range(pivot,len(lst)):
        lst[i] = 0
    return lst

path = input()
print(moveZeroes(path))

```

```python
from math import pi
class Square:
    def __init__(self,side):
        self.side = side
    def perimeter(self):
        return self.side * 4

class Rectangle:
    def __init__(self,width,height):
        self.width = width
        self.height = height
    def perimeter(self):
        return (self.width + self.height) * 2
class Circle:
    def __init__(self,radius):
        self.radius = radius
    def perimeter(self):
        return 2 * self.radius * pi

def petri(path:str):
  with open(path, "r") as f:
      infor = "Type"
      for line in f:
          if line == "\n":
              infor = "Type"
              continue

          if line[:4] == "TYPE":
              line = line.rstrip().split(": ")
              infor = line[1]
              print(infor)
          else:
              if infor == "SQUARE":
                  line = line.rstrip().split(": ")
                  side = float(line[1])
                  s = Square(side)
              elif infor == "RECTANGLE":
                  a,b,c,d = line.split()
                  b,d = float(b), float(d)
                  s = Rectangle(b, d)
              else:
                  line = line.rstrip().split(": ")
                  side = float(line[1])
                  s = Circle(side)

              print(s.perimeter())
path = input()
petri(path)
```

```python
def imgSmoother(path:str):
    with open(path, "r") as f:
        M = int(f.readline())
        matrix = []
        for i in range(M):
            line = f.readline()
            line = list(map(int,line.split()))
            matrix.append(line)

    result = [[0] * M for _ in range(M)]

    for levels in range(M):
        for idx in range(M):
            count = 1
            for x,y in zip([1,1,1,0,0,-1,-1,-1],[1,-1,0,1,-1,1,-1,0]):
                if (levels + x < 0 or levels + x >= M) or (idx + y < 0 or idx + y >= M):
                    continue
                result[levels][idx] += matrix[levels + x][idx + y]
                count += 1
            result[levels][idx] = (result[levels][idx] + matrix[levels][idx]) // count
    return result
path = input()
print(imgSmoother(path))
```
