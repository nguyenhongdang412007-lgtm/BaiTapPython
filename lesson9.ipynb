```python
def merge_sort(arr):
    if len(arr)<=1:
        return arr
    else:
        mid=len(arr)//2
        left=merge_sort(arr[:mid])
        right=merge_sort(arr[mid:])
    return merge(left,right)

def merge(left,right):
    res=[]
    i=j=0
    
    while i<len(left) and j<len(right):
        if left[i] <=right[j]:
            res.append(left[i])
            i+=1
        else:
            res.append(right[j])
            j+=1
    res.extend(left[i:])
    res.extend(right[j:])
    return res

a=list(map(int,input().split()))
b=list(map(int,input().split()))
c=a+b
for x in merge_sort(c):
    print(f'{x}',end=" ")
```

```python
def count_leq(matrix, x):
    n = len(matrix)
    m = len(matrix[0])

    i = n - 1      
    j = 0
    count = 0
    while i >= 0 and j < m:
        if matrix[i][j] <= x:
            count += i + 1   
            j += 1
        else:
            i -= 1
    return count


def kth_largest(matrix, k):
    n = len(matrix)
    m = len(matrix[0])
    total = n * m

    k_small = total - k + 1

    low = matrix[0][0]
    high = matrix[n - 1][m - 1]

    while low < high:
        mid = (low + high) // 2
        if count_leq(matrix, mid) < k_small:
            low = mid + 1
        else:
            high = mid

    return low

n, m = map(int, input().split())

matrix = []
for _ in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)
k=int(input())
print("Số lớn thứ", k, "là:", kth_largest(matrix, k))
```

```python
def w9a6(a):
    n = len(a)
    i = n - 2
    while i >= 0 and a[i] >= a[i + 1]:
        i -= 1

    if i < 0:
        a.reverse()
        return a

    j = n - 1
    while a[j] <= a[i]:
        j -= 1
    a[i], a[j] = a[j], a[i]
    a[i + 1:] = reversed(a[i + 1:])
    return a

n = int(input())
a = list(map(int, input().split()))
res = w9a6(a)

print(*res) #unpacking
```

```python
def canPlaceFlowers(flowerbed,k):
    count=0
    if len(flowerbed) <=1:
        if flowerbed[0] ==0:
            count+=1
    else:
        if flowerbed[0]==0 and flowerbed[1]==0:
            flowerbed[0]=1
            count+=1
        if flowerbed[len(flowerbed)-1]==0 and flowerbed[len(flowerbed)-2]==0:
            flowerbed[len(flowerbed)-1]=1
            count+=1
        for i in range(1,len(flowerbed)-1):
            if flowerbed[i-1] ==0 and flowerbed[i+1]==0 and flowerbed[i]==0:
                count+=1
    if count>=k:
        return True
    return False 

a=list(map(int,input().split()))
k=int(input())
if canPlaceFlowers(a,k):
    print("True")
else:
    print("False")

    
    
```

```python
def is_path_crossing(moves):
    x,y=0,0
    d={}
    for a in moves:
        if a=="R":
            x+=1
            if (x,y) not in d:
                d[(x,y)] =1
            else:
                d[(x,y)]+=1
        if a=="L":
            x-=1
            if (x,y) not in d:
                d[(x,y)] =1
            else:
                d[(x,y)]+=1
        if a=="D":
            y-=1
            if (x,y) not in d:
                d[(x,y)] =1
            else:
                d[(x,y)]+=1
        if a=="U":
            y+=1
            if (x,y) not in d:
                d[(x,y)] =1
            else:
                d[(x,y)]+=1
    for z in d:
        if d[z]>1 or x==y==0:
            return True
    return False

a=input()
print(is_path_crossing(a))

```

```python
def Median(nums):
    a, b, c, d, e = nums

    if a > b:
        a, b = b, a

    if c > d:
        c, d = d, c

    if a > c:
        a, c = c, a

    if b > d:
        b, d = d, b

    if b > c:
        b, c = c, b

    if e < b:
        return b
    elif e > c:
        return c
    else:
        return e


a=list(map(int,input().split()))
Median(a)
    
```

```python
def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()

    j = 0
    res = 0

    for h in houses:
        while j + 1 < len(heaters) and abs(heaters[j + 1] - h) <= abs(heaters[j] - h):
            j += 1
        res = max(res, abs(heaters[j] - h))

    return res
houses = list(map(int,input().split()))
heaters = list(map(int,input().split()))
print(findRadius(houses, heaters))
```
