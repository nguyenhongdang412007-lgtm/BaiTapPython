```python
# W6A4
def chia(s):
    for i in range(len(s)):
        if s[i] == ':':
            return (s[:i], s[i + 1:])
a = input().split()
d = {}
for x, y in [chia(x) for x in a]:
    if x in d:
        d[x].append(y)
    else:
        d[x] = [y]
print(d)
```

```python
# W6A5
a = list(map(int, input().split()))
d = {'positives': 0, 'negatives': 0, 'zero':0}
for x in a:
    if x == 0:
        d['zero'] += 1
    elif x < 0:
        d['negatives'] += 1
    else:
        d['positives'] += 1
print(d)
```

```python
# W6A10
a = list(map(int, input().split()))
k = int(input())
res = []
for i in range(len(a)):
    for j in range(i + 1, len(a)):
        if a[i] + a[j] == k:
            res.append((a[i], a[j]))
            break
res = list(set(res))
res.sort()
print(res)
```

```python
# W6A12
a = list(map(int, input().split()))
d = {}
res = (0, 0)
for x in a:
    if x not in d:
        d[x] = 1
    else:
        d[x] += 1
    res = max(res, (d[x], -x))
print(-res[1])
```

```python
# W6A19
def cal(a):
    for i in range(len(a)):
        if a[i] == max(a):
            return i + 1
    return -1
a = list(map(int, input().split()))
print(cal(a))
```

```python
# W6A20
def check(a, k):
    for i in range(1, len(a)):
        if a[i - 1] == k:
            return i
    return -1
a = list(map(int, input().split()))
k = int(input())
print(check(a, k))
```
