```python
#w5a1
def max_of_two_nums(a,b):
    return a if a>b else b
a,b=map(int,input().split())
print (max_of_two_nums(a,b))
```

```python
#w5a2
def swap(a,b):
    return b,a
a,b=map(int,input().split())
a,b=swap(a,b)
print(a,b)
```

```python
#w5a3
def isPrime(n):
    for i in range (2,int(n**0.5)+1):
        if n%i==0:
            return False
    return n>1
n=int(input())
print(isPrime(n))
    
```

```python
#w5a4
def isPerfect(n):
    total=0
    for i in range(1,n):
        if n%i==0:
            total+=i
    return total==n
n=int(input())
print(isPerfect(n))
```

```python
#w5a5
def index(l,target):
    for i in range(len(l)):
        if l[i]==target:
            return i+1
    return -1
l=list(map(int,input().split()))
target=int(input())
print(index(l,target))
```

```python
#w5a6
def fac(n):
    res=1
    for i in range(1,n+1):
        res*=i
    return res
n=int(input())
print(fac(n))
```

```python
#w5a7
def isvalid(s):
    n=len(s)
    if not s[n-1].isdigit(): return False
    for i in range(n):
        if (s[i]=='+' or s[i]=='-' or s[i]=='*' or s[i]=='/') and not s[i+1].isdigit():
            return False
        elif s[i]=='/' and float(s[i+1:])==0: return False
    return True
def mycalculator(s):
    n=len(s)
    for i in range(n):
        if s[i]=='+': 
            return float(s[:i]) + float(s[i+1:])
        elif s[i]=='-': 
            return float(s[:i]) - float(s[i+1:])
        elif s[i]=='*': 
            return float(s[:i]) * float(s[i+1:])
        elif s[i]=='/': 
            return float(s[:i]) / float(s[i+1:]) 
s=input()
print(f"{mycalculator(s):.2f}") if isvalid(s) else print("INVALID!!!!!!")
        
```

```python
#w5a8
def hamming(x,y):
    a=x^y
    res=0
    while a>0:
        if a%2!=0:
            res+=1
        a//=2
    return res
x,y=map(int,input().split())
print(hamming(x,y))
```

```python
#w5a9
def sumofnum(n):
    s=str(n)
    sum=0
    for i in s:
        sum+=int(i)
    return sum
n=int(input())
print(sumofnum(n))
```

```python
#w5a10
def mapping(a,b):
    d=dict()
    for i in range(len(a)):
        if d.get(a[i]) and b[i]!=d[a[i]]:
            return False
        d[a[i]]=b[i]
    return True
a,b=map(str,input().split())
print(mapping(a,b) and mapping(b,a))
```
