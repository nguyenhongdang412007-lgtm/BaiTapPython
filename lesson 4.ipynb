#W4A26

```python
def fibo(n):
    if n==1:
        return 1
    if n==0:
        return 0
    else:
        return fibo(n-1)+fibo(n-2)

A=int(input())
n=0
while fibo(n)<A:
    n+=1
print(fibo(n-1))
```
